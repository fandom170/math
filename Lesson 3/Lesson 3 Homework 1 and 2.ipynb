Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)  
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
A_ = [10, 10, 10]
B_ = [0, 0, 10]
```

```python
def vector_sum(A_, B_):
    if len(A_) != len(B_):
        print("Vectors have different dimensions, please check conditions")
    else:
        total = 0
        C_ = []
        for i in range(len(A_)):
            C_.append(A_[i] + B_[i])
            total += (A_[i] - B_[i])**2
        print(f'координаты суммирующего вектора для  векторов А и В {C_}')
        print(f'длина результирующего вектора для векторов А и В {round(np.sqrt(total), 2)}')
```

```python
vector_sum(A_, B_)
```

*Примечание*  
Указанное решение выше справедливо, если предполагается, что вектора выше лежат в одной плоскости и угол между ними 90 градусов (являются результатом разложения исходного вектора).

```python
def draw_parabola():
    x = np.linspace(-20, 20, 1000)
    y = np.linspace(-20, 20, 1000)
    x, y = np.meshgrid(x, y)
    print('Please enter p')
    p=2
    p = float(input())
    if p<=0:
        print("Parabola cannot be built. Please enter positive p")       
        return None
    
    plt.gca().set_aspect('equal', adjustable='box')
    axes()
    plt.contour(x, y, (y**2 - 4*p*x), [0], colors='k')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
def draw_circle():
    """Fraws circle or ellipse fepends on koefficients"""
    print('Please enter a and b')
    a = 1 
    b = 1
    a, b = map(float,(input().split()))
       
    if a<=0 or b <=0:
        print("Circle or ellipse cannot be built. Please enter positive numbers")       
        return None
    plt.gca().set_aspect('equal', adjustable='box')
    axes()
    x = np.linspace(-2*a, 2*a, 1000)
    y = np.linspace(-2*b, 2*b, 1000)
    x, y = np.meshgrid(x, y)
    plt.contour(x, y, (x**2/a**2 + y**2/b**2), [1], colors='k')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
def draw_hyperbola():
    """Fraws circle or ellipse fepends on koefficients"""
    print('Please enter a and b')
    a = 1 
    b = 1
    a, b = map(float,(input().split()))
    
    if a<=0 or b <=0:
        print("Circle or ellipse cannot be built. Please enter positive numbers")       
        return None
    
    plt.gca().set_aspect('equal', adjustable='box')
    axes()
    x = np.linspace(-5*a, 5*a, 1000)
    y = np.linspace(-5*b, 5*b, 1000)
    x, y = np.meshgrid(x, y)
    plt.contour(x, y, (x**2/a**2 - y**2/b**2), [1], colors='k')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
print('Please enter desired curve type and koeficionts. Available variants are:\n -parabola\n -circle\n -hyperbola')
type = input().lower()
if type=='parabola':
    draw_parabola()
elif type=='hyperbola':
    draw_hyperbola()
elif type=='circle' or type=='ellipse':
    draw_circle()
else:
    print('Unknown curve')
```

Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.  

```python
#x = r*cos(theta)
#y = r = sin(theta)
# ax + by + c = 0

def polar_line(a, b, c, angle=np.pi):
    plt.axes(projection='polar')
    angles = np.arange(0, angle, 0.01)
    for theta in angles:
        r = c/(a*np.cos(theta) + b * np.sin(theta))
        plt.polar(theta, r, 'r.')
    plt.grid(True)
    plt.show()   
    
```

```python
polar_line(1, 2, -1, np.pi*0.5)
```

Решите систему уравнений
exp(x) + x∙(1 – y) = 1   
y = x2 – 1  





```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-5, 5, 250)
plt.plot(x, x**2-1, color='orange', label='x**2-1')
plt.plot(x, 1-((1-exp(x))/(x)),  color='blue', label='1-((1-exp(x))/(x))')
plt.hlines(0, -5, 5, color='green', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('True')
plt.legend(loc='upper left')
plt.show()
```

Преобразуем уравнения, чтобы в правой части был ноль.  
exp(x) + x∙(1 – y) - 1= 0  
x^2 – 1 - y = 0

```python
def equations(p):
    x, y = p
    return ((1-(1-exp(x))/x)-y, x**2-1-y)
```

```python
x = []
y = []
for i in range (-3, 6, 1):        
    if i == 0:
        continue
    x1, y1 = fsolve(equations, (i, i))
    x1 = round(x1, 7)
    y1 = round(y1, 7)
    if (x1 not in x or y1 not in y):
        x.append(x1)
        y.append(y1) 
    
    
```

```python
print("Корни системы уравнений:")
for i in range(len(x)):
    print(f'X = {x[i]}, Y = {y[i]}')
```
