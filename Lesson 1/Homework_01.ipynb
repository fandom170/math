```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
  %matplotlib inline
```

```python
x = np.linspace(-10, 10, 200)
```

```python
k1 = 1
k2 = np.pi
```

```python
figure(figsize=(12, 8), dpi=80)
plt.plot(x, np.cos(k1*x), "-b", label = f"cos(k * x), k = {k1}")
plt.plot(x, np.cos(k2*x), "-r", label = f"cos(k * x), k = {k2}")
plt.legend(loc="upper left")
plt.show()
```
