#### Численный метод решения

```python
def x_calc(k):
    a_list = []
    a = np.linspace(0.01, 0.02, 10000)
    x_pred = 0
    for a_ in a[::-1]:
        x = np.pi * k / a_
        if x >= 500 or x <= abs(100):
            break
        else:
            x_pred = x
            a_list.append(a_)
    return a_list
    
        
```

```python
for k in range(-10, 10, 1):
    a_list = x_calc(k)
    if len(a_list) > 0:
        print(f'{a_list[0]}, {a_list[-1]}, k = {k} ')
        
```

Таким образом, существуют три возможные ветки решения при различных k:  

    - (0.02 , 0.01) при к = 1  
    - (0.02 , 0.01257) при к = 2  
    - (0.02, 0.1885) при к = 3
